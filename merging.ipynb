```python
# Creating the extended bell schedule DataFrame
extended_bell_schedule_df = pd.DataFrame(full_bell_schedule_data)

# Ensuring 'Prep' column in teacher_prep_df is treated as strings and handling missing values
teacher_prep_df['Prep'] = teacher_prep_df['Prep'].fillna('').astype(str)

# Merging the extended datasets based on periods in the schedule
extended_combined_data = []
for _, row in extended_bell_schedule_df.iterrows():
    date = row['DATE']
    schedule = row['SCHEDULE']
    for _, teacher_row in teacher_prep_df.iterrows():
        teacher_name = teacher_row['Name']
        prep_periods = teacher_row['Prep'].split(', ') if teacher_row['Prep'] else []
        prep = 'No'
        if schedule != 'No School' and prep_periods:
            for period in prep_periods:
                if period in schedule:
                    prep = 'Yes'
                    break
        elif schedule == 'No School':
            prep = 'No School'
        extended_combined_data.append({
            'DATE': date,
            'Teacher Name': teacher_name,
            'Prep': teacher_row['Prep'],
            'SCHEDULE': schedule,
            'Has Prep': prep
        })

# Creating the extended combined DataFrame
extended_combined_df = pd.DataFrame(extended_combined_data)

# Displaying the extended combined DataFrame
extended_combined_df
```

```python

final_df_cleaned = extended_combined_df[extended_combined_df['SCHEDULE'] != 'No School']

# Also remove rows where either 'Teacher Name' or 'Prep' column has NaN values
final_df_cleaned = final_df_cleaned.dropna(subset=['Teacher Name', 'Prep'])

# Save the cleaned DataFrame to an Excel file
file_save_path = 'Extended_Combined_Teacher_Schedule_Cleaned_2.xlsx'
final_df_cleaned.to_excel(file_save_path, index=False)

# Download the Excel file
from google.colab import files
files.download(file_save_path)

```
